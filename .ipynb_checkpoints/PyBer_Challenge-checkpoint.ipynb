# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
all_cities_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
all_cities_df
```

```python
#urban_cities_df.date = pd.to_datetime(urban_cities_df.date)
all_cities_df.info()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
# Get the number of rides for urban cities.
urban_ride_total = all_cities_df.groupby(["type"]).count()["ride_id"]
urban_ride_total
```

### This is where I dropped duplicate city values to return accurate driver count.

```python
#urban_cities_df[all_cities_df.city.eq("West Angela")]
#create dropped duplicates dataframe 
unique_cities_df = all_cities_df.drop_duplicates(["city"])
```

```python
#city value count
unique_cities_df.city.value_counts()
```

```python
# 2. Get the total drivers for each city type
city_type_driver_total = all_cities_df.groupby(["type"]).sum()["driver_count"]
city_type_driver_total

# total_drivers_by_type = city_data_df.groupby(["type"]).sum()["driver_count"]
# total_drivers_by_type
#city_type_driver_total = urban_cities_df.groupby(urban_cities_df["type"]).sum()["driver_count"]
#city_type_driver_total
```

```python
#  3. Get the total amount of fares for each city type
fares_city_type_total = all_cities_df.groupby(["type"])["fare"].sum()
fares_city_type_total
```

```python
# # urban city DataFrame.
# urban_cities_df = urban_cities_df[urban_cities_df["type"] == "Urban"]
# rural_cities_df = urban_cities_df[urban_cities_df["type"] == "Rural"]
# suburban_cities_df = urban_cities_df[urban_cities_df["type"] == "Suburban"]
# #urban_cities_df.head()
# #rural_cities_df.head()
# #suburban_cities_df.head()
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_perride_city_type = fares_city_type_total/urban_ride_total

avg_fare_perride_city_type
```

```python
# 5. Get the average fare per driver for each city type. 
# avg_fare_perdriver_city_type = fares_city_type_total/city_type_driver_total
avg_fare_perdriver_city_type = fares_city_type_total/city_type_driver_total
avg_fare_perdriver_city_type
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({
         "Total Rides City Type": urban_ride_total,
         "Total Drivers City Type": city_type_driver_total,
         "Total Fares City Type": fares_city_type_total,
         "Average Fare Per Ride City Type": avg_fare_perride_city_type,
         "Average Fare Per Driver City Type" : avg_fare_perdriver_city_type})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
# Format the "Total fares City Type" to have the comma for a thousands separator, a decimal separator and a "$".
pyber_summary_df["Total Rides City Type"] = pyber_summary_df["Total Rides City Type"].map("{:,}".format)
pyber_summary_df["Total Drivers City Type"] = pyber_summary_df["Total Drivers City Type"].map("{:,}".format)
pyber_summary_df["Total Fares City Type"] = pyber_summary_df["Total Fares City Type"].map("${:,.2f}".format)
pyber_summary_df["Average Fare Per Ride City Type"] = pyber_summary_df["Average Fare Per Ride City Type"].map("${:,.2f}".format)
pyber_summary_df["Average Fare Per Driver City Type"] = pyber_summary_df["Average Fare Per Driver City Type"].map("${:,.2f}".format)

# Display the data frame
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
# pyber_summary_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
fare_total_date_df = pd.DataFrame(all_cities_df.groupby(["type", "date"]).sum()["fare"])
fare_total_date_df
```

### Note: this is where I removed duplicates

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
fare_total_date_df = fare_total_date_df.reset_index()
fare_total_date_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
fare_date_pivot_table_df = fare_total_date_df.pivot(index = "date", columns = "type", values = "fare") 
fare_date_pivot_table_df

```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
date_range_df = fare_date_pivot_table_df.loc["2019-01-01":"2019-04-29"]
date_range_df

```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.

# df.index = pd.to_datetime(df.index)
date_range_df.index = pd.to_datetime(date_range_df.index) 

date_range_df

```

```python
# 7. Check that the datatype for the index is datetime using df.info()
date_range_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
per_week_df = date_range_df.resample("w").sum()
per_week_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
per_week_df.plot(
    figsize = (20,6),
    title = "Total Fare by City Type",
    xlabel = "Time",
    ylabel = "Fare($USD)"
)
plt.savefig("Resources/Pyber_fare_summary.png")
```
